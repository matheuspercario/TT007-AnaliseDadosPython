# Listas e tuplas em Python

Estruturas de dados usadas para armazenar itens de dados relacionados em Python são chamadas **sequências**.

- Python suporta três tipos básicos de sequencias: a **string**, a **lista** e a **tupla**. 

- Mapeamentos (Mappings), também chamados de vetor associativo ou tabela hash, são estruturas de dados que armazenam dados usando uma chave.  
    - Em Python existe um tipo de mapeamento, chamado dicionário.


Uma sequência é uma série de valores <font color=blue >relacionados e subsequentes na memória </font>.
Em Python, sequências podem ser acessadas pelo final.
- Último elemento é acessado pelo indice -1.
- len(c) – retorna o tamanho da sequência.

As **listas geralmente armazenam dados homogêneos**, ou seja, valores do mesmo tipo de dados. Considere o lista c, que contém cinco elementos inteiros:


```python
c = [-45, 6, 0, 72, 1543]
print(c)
```

### Alocação dinâmica
Vamos começar com uma lista vazia [] e, em seguida, usar uma instrução _for_ e **+=** para acrescentar os valores de 1 até 5 na lista - a lista cresce dinamicamente para acomodar cada item:



```python
a_list = []  # cria lista vazia
for number in range(1, 6):
    a_list += [number]    # insere elementos na lista

print(a_list)
```

Quando o operando esquerdo de **+=** é uma lista, o operando direito deve ser **iterável**; caso contrário, um **TypeError** ocorre. 
Se o operando à direita de **+=** contém vários elementos, todos são acrescentados à lista. 

```python
letters = []
letters += 'Python'
print(letters)
```

### Concatenando com +
Você pode concatenar duas listas, duas tuplas ou duas strings usando o operador **+**. O resultado é uma **nova sequência do mesmo tipo** contendo os elementos do operando esquerdo seguidos pelo elementos do operando direito. *As sequências originais não são alteradas*:


```python
list1 = [10, 20, 30]
list2 = [40, 50]
concatenated_list = list1 + list2   # concatena listas
print(concatenated_list)
```

#### Acessando elementos da lista com _for_ e _range_
Os elementos da lista também podem ser acessados por meio de seus índices e do operador de assinatura ([]):


```python
for i in range(len(concatenated_list)):
    print(f'{i}: {concatenated_list[i]}')
```

A chamda da função *range* com o argumento *(len(concatenated_list))* produz uma sequência de inteiros que representam índices de *concatenated_list* (neste caso, 0 a 4).

#### Operadores de comparação
Você pode comparar listas inteiras, elemento por elemento, usando operadores de comparação:


```python
a = [1, 2, 3]
b = [1, 2, 3]
c = [1, 2, 3, 4]
print(f'{a} == {b} is {a == b}') # True: corresponding elements in both are equal
print(f'{a} == {c} is {a == c}') # False: a and c have different elements and lengths
print(f'{a} < {c} is {a < c}') # True: a has fewer elements than c
print(f'{c} >= {b} is {c >= b}')# True: elements 0-2 are equal but c has more elements
```

#### Acessando elementos em uma tupla

**Os elementos de uma tupla, embora relacionados, são frequentemente de vários tipos**. Normalmente, você não itera sobre eles. Em vez disso, você acessa cada um individualmente. Como os índices de lista, os índices de tupla começam em 0

```python
hour = int( input('Enter hour:') )
minute = int( input('Enter minute:') )
second = int( input('Enter second:') )

currentTime = hour, minute, second # cria tupla

print("The value of currentTime is:", currentTime)

# accessa a tupla
print("The number of seconds since midnight is", \
( currentTime[ 0 ] * 3600 + currentTime[ 1 ] * 60 + currentTime[ 2 ] ))
```

#### Adicionando itens a uma tupla

Tal como acontece com as listas, a instrução de atribuição **+=** pode ser usada com strings e tuplas, mesmo que eles sejam imutáveis.
No código a seguir, após as duas atribuições, _tupla1_ e _tupla2_ referem-se ao mesmo objeto de tupla:


```python
tuple1 = (10, 20, 30)
tuple2 = tuple1
tuple2
```

```python
tuple1 += (40, 50)
tuple1
```

```python
tuple2
```

Concatenar a tupla (40, 50) para _tupla1_ cria uma nova tupla e, em seguida, atribui uma referência para a variável *tupla1*; *tupla2* ainda se refere à tupla original

Para uma string ou tupla, o item à direita de **+=** deve ser uma string ou tupla, respectivamente. A mistura de tipos causa um **TypeError**.
No entanto, você pode usar **+=** para anexar uma tupla a uma lista!

```python
numbers = [1, 2, 3, 4, 5]
numbers += (6,7)
numbers
```

#### Acessando índices e valores com segurança com a função _enumerate_

Anteriormente, chamamos _range_ para produzir uma sequência de valores de índice e, em seguida, acessamos os elementos da lista em um laço _for_ usando os valores de índice e o operador ([]). 
- Isso é sujeito a erros porque você pode passar os **argumentos errados** para *range*. Se algum valor produzido por _range_ for um índice fora dos limites, usá-lo como um índice causa um **IndexError**. 

O mecanismo preferido para acessar o índice e o valor de um elemento é a função **enumerate**

```python
colors = ['red','green','blue']
for indice, valor in enumerate(colors):
    print(f'{indice}: {valor}')
```

Esta função recebe um iterável e cria um **iterador** que, para cada elemento, retorna uma **tupla contendo o índice e o valor do elemento**.

```python
list(enumerate(colors))
```

Função **tuple** cria uma tupla de uma sequência


```python
tuple(enumerate(colors))
```

### Passando listas para funções

Como vimos, todos os objetos são passados por referência a uma função. 
Considere a função *modify_elements*, que recebe uma referência a uma lista e multiplica cada um dos valores de elemento da lista por 2:


```python
def modify_elements(items):
    """"Multiplies all element values in items by 2."""
    for i in range(len(items)):
        items[i] *= 2
```

```python
numbers = [10, 3, 7, 1, 9]
modify_elements(numbers)
numbers
```

Uma expressão geradora é semelhante a uma notação simplificada de lista, mas cria um gerador iterável que produz valores sob demanda. Isso é conhecido como avaliação preguiçosa. 
- Notação simplificada de lista usa avaliação gananciosa - eles criam listas imediatamente quando você as executa. 
- Para um grande número de itens, a criação de uma lista pode consumir bastante memória e tempo. Então gerador expressões podem reduzir o consumo de memória do seu programa e melhorar o desempenho se a lista inteira não é necessária de uma vez.

As expressões geradoras têm os mesmos recursos que a notação simplificada de lista. Você
pode defina-las entre parênteses em vez de colchetes.

```python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
for value in (x ** 2 for x in numbers if x % 2 != 0):
    print(value, end=' ')
```

```python
list((x ** 2 for x in numbers if x % 2 != 0))
```

Para mostrar que uma expressão geradora não cria uma lista, vamos atribuir a expressão anterior
 para uma variável e avaliar a variável:

```python
squares_of_odds = (x ** 2 for x in numbers if x % 2 != 0)
squares_of_odds
```

### Filtrar, mapear e reduzir

Aqui, demonstramos o filtro integrado e as funções de mapa para filtrar
e mapeamento, respectivamente. Continuamos discutindo as reduções em que você processa
uma coleção de elementos em um único valor, como sua contagem, total, produto, média,
mínimo ou máximo.

Vamos usar a função _filter_ para obter os valores ímpares em numbers:

```python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

def is_odd(x):
    """Returns True only if x is odd."""
    return x % 2 != 0

list(filter(is_odd, numbers)) # a função list itera sobre os resultados para criar a lista
```

Assim como dados, funções em Python são objetos que você pode atribuir a variáveis, passar para 
outras funções e retornar de funções. 
- Funções que recebem outras funções como argumentos são chamadas funções de ordem superior. 

Por exemplo, o primeiro argumento de **filter** deve ser uma função que recebe um argumento e retorna True se o valor deve ser incluído no resultado.
A função **is_odd** retorna True se seu argumento for ímpar.
A função de **filter** chama **is_odd** <font color=red >uma vez para cada valor de seu segundo argumento iterável (number) </font>. 
As funções de ordem superior também podem retornar uma função como resultado.

A função **filter** retorna um iterador, então os resultados do filtro não são produzidos até que você iterar por meio deles. Este é outro exemplo de avaliação preguiçosa.

#### Usando lambda ao invés de função
Para funções simples como **is_odd** que retornam apenas o valor de uma única expressão, você pode usar um expressão lambda (ou simplesmente lambda) para definir a função embutida onde é necessária—normalmente quando é passado para outra função:

```python
list(filter(lambda x: x % 2 != 0, numbers))
```

Passamos o valor de retorno de **filter** (um iterador) para a função **list** para converter 
os resultados para uma lista e exibí-los.
Uma expressão <font color=blue> lambda é uma função anônima </font> - ou seja, uma função sem nome. 

Na chamada de filter

**<center> filter(lambda x: x % 2! = 0, numbers) </center>**

o primeiro argumento é lambda

**<center> lambda x: x % 2! = 0 </center>**

lambda começa com a palavra-chave lambda seguida por uma lista de parâmetros separados por vírgulas, dois pontos (:) e uma expressão. Nesse caso, a lista de parâmetros possui um único parâmetro, x.

lambda retorna implicitamente o valor de sua expressão.

Mapeando os valores de uma sequência para novos valores usando a função _**map**_

```python
numbers
```

```python
list(map(lambda x: x ** 2, numbers))
```

O primeiro argumento do mapa de funções é uma função que recebe um valor e retorna um novo
valor — neste caso, um lambda que quadrada seu argumento. O segundo argumento é um iterável
de valores para mapear. O mapa de funções usa avaliação preguiçosa. Então, passamos para a função de lista o
iterador que o mapa retorna. Isso nos permite iterar e criar uma lista dos mapeados
valores. Aqui está uma compreensão de lista equivalente:

```python
[item ** 2 for item in numbers]
```

**Combinando _map_ e _filter_**

```python
list(map(lambda x: x ** 2, filter(lambda x: x % 2 != 0, numbers)))
```

Muita coisa está acontecendo na linha anterior. Primeiro, **filter** retorna um iterável que representa apenas os valores ímpares dos números. Em seguida, **map** retorna um iterável que representa os quadrados dos valores filtrados. Finalmente, **list** usa o iterável de **map** para criar o lista. Você pode preferir a notação simplificada..

```python
[x ** 2 for x in numbers if x % 2 != 0]
```

#### Combinando iteráveis em tuplas de elementos correspondentes

A função **zip** permite iterar vários iteráveis de dados ao mesmo tempo.
A função recebe como argumentos qualquer número de iteráveis e retorna um iterador que
produz tuplas contendo os elementos no mesmo índice em cada um.

```python
names = ['Bob', 'Sue', 'Amanda']
grade_point_averages = [3.5, 4.0,]

for name, gpa in zip(names, grade_point_averages):
    print(f'Name={name}; GPA={gpa}')
```

No exemplo **zip** produz as tuplas: ('Bob', 3.5), ('Sue', 4.0) e ('Amanda', 3,75)
consistindo nos elementos no índice 0, 1 e 2 de cada lista, respectivamente.

No laço for as tuplas são desempacotadas em _name_ e _gpa_ e seus valores são exibidos. 

<font color=blue>O argumento mais curto da função **zip** determina o número de tuplas 
produzidas </font>. 

Aqui, ambos têm o mesmo comprimento.

### Listas bidimensinais

As listas podem conter outras listas como elementos. Um uso típico é representar tabelas de valores que consistem em informações organizadas em linhas e colunas.

- Listas multidimensionais podem ter mais de dois índices. Aqui, apresentamos listas bidimensionais.

### Visualização de frequências e porcentagens de jogadas de dados

```python
# RollDie.py
"""Graphing frequencies of die rolls with Seaborn."""
import matplotlib.pyplot as plt
import numpy as np
import random 
import seaborn as sns
import sys

# use list comprehension to create a list of rolls of a six-sided die
rolls = [random.randrange(1, 7) for i in range(60000)]

# NumPy unique function returns unique faces and frequency of each face
values, frequencies = np.unique(rolls, return_counts=True)

title = f'Rolling a Six-Sided Die {len(rolls):,} Times'
sns.set_style('whitegrid')  # white backround with gray grid lines
axes = sns.barplot(values, frequencies, palette='bright')  # create bars
axes.set_title(title)  # set graph title
axes.set(xlabel='Die Value', ylabel='Frequency')  # label the axes

# scale y-axis by 10% to make room for text above bars
axes.set_ylim(top=max(frequencies) * 1.10)

# display frequency & percentage above each patch (bar)
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0  
    text_y = bar.get_height() 
    text = f'{frequency:,}\n{frequency / len(rolls):.3%}'
    axes.text(text_x, text_y, text, 
              fontsize=11, ha='center', va='bottom')

plt.show()  # display graph 
```

1. O módulo **matplotlib.pyplot** contém os recursos gráficos da biblioteca **Matplotlib** que usamos. 
    - Este módulo normalmente é importado com o nome plt.
2. A biblioteca **NumPy** (Numerical Python) inclui a função **unique** que iremos use para resumir as jogadas de dados. 
    - O módulo numpy normalmente é importado como np.
3. O módulo **random** contém as funções de geração de números aleatórios do Python.
4. O módulo **seaborn** contém os recursos gráficos da biblioteca **Seaborn** que usamos. 
    - Este módulo normalmente é importado com o nome sns.

### Criando o gráfico de barra inicial

A f-string inclui o número de jogadas de dados no título do gráfico da barra. A vírgula

### Definindo o título da janela e rotular os eixos x e y

O método **set_title** do objeto de **Axes** é usado para exibir a string de título centralizada. Este método retorna um objeto **Text** contendo o título e sua localização

#### Finalizando o gráfico de barras
Os próximos dois trechos completam o gráfico abrindo espaço para o texto acima de cada barra, exibindo-o em seguida.

Para abrir espaço para o texto acima das barras, a primeira linha dimensiona o eixo y em 10% - valor escolhido por meio de experimentação. 
- O método **set_ylim** do objeto Axes tem muitos argumentos de palavra-chave opcionais. 
    - Aqui, usamos apenas **top** para alterar o valor máximo representado pelo eixo y.
    - Multiplicamos a maior frequência por 1,10 para garantir que o eixo y é 10% mais alta que a barra mais alta.

Finalmente, o valor de frequência de cada barra e a porcentagem total dos valores dos dados é exibida. 
- O objeto **patches** de Axes contém formas coloridas bidimensionais que representam as barras do gráfico. 

As instruções _for_ usa _zip_ são usadas para percorrer os patches e seus valores de frequência correspondentes. 
- Cada iteração descompacta uma barra e frequência de uma das tuplas que zip retorna. 
- O conjunto de instruções *for* opera da seguinte forma:
    - A primeira instrução calcula a coordenada x central onde o texto aparecerá.  
        - Calculamos isso como a soma da coordenada x da borda esquerda da barra (bar.get_x ()) e metade da largura da barra (bar.get_width () / 2.0).
    - A segunda declaração obtém a coordenada y onde o texto aparecerá bar.get_y () representa o topo da barra. 
    - A terceira declaração cria uma string de duas linhas contendo a frequência dessa barra e a porcentagem correspondente do total de jogadas de dados.
    - A última instrução chama o método de texto do objeto Axes para exibir o texto acima de bar. Os dois primeiros argumentos deste método especificam a posição x – y do texto, e o  terceiro argumento é o texto a ser exibido. 
        - O argumento de palavra-chave **ha** especifica o horizontal alinhamento - centralizamos o texto horizontalmente em torno da coordenada x. O argumento de palavra-chave **va** especifica o alinhamento vertical - alinhamos a parte inferior de o texto com na coordenada y.
